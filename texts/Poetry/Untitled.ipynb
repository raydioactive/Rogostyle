```python
import os
import re

# Directory containing the text files
directory = "/slipstream_old/home/juliusherzog/Rogostyle/texts/Poetry"


# Regular expression pattern to match non-Latin alphabet characters and Arabic numbers (excluding periods)
pattern = r'[^a-zA-Z\s\.]'

# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(directory, filename)

        # Read the contents of the file
        with open(file_path, 'r') as file:
            text = file.read()

        # Remove non-Latin alphabet characters and Arabic numbers (excluding periods) using regex
        processed_text = re.sub(pattern, '', text)

        # Remove remaining Arabic numbers
        processed_text = ''.join(char for char in processed_text if not char.isdigit())

        # Write the processed text back to the file
        with open(file_path, 'w') as file:
            file.write(processed_text)

        print(f"Processed {filename}")

print("Preprocessing completed.")
```

```python
import os
import shutil
from cltkreaders.lat import LatinLibraryCorpusReader

# Create an instance of the LatinLibraryCorpusReader
lib = LatinLibraryCorpusReader()

# Destination directory
dest_dir = os.path.expanduser("~/Rogostyle/texts/Poetry")

# Create the destination directory if it doesn't exist
os.makedirs(dest_dir, exist_ok=True)

# List of poetry directories
poetry_dirs = [
    "horace",
    "juvenal",
    "lucretius",
    "martial",
    "ovid",
    "prudentius",
    "statius",
    "vergil"
]

# Recursively move poetry texts from directories
def move_poetry_texts(directory):
    try:
        # Get the list of file IDs in the directory
        file_ids = lib.fileids(directory)
        
        for file_id in file_ids:
            file_path = f"{directory}/{file_id}"
            
            if file_id.endswith(".txt"):
                # Move the text file to the destination directory
                shutil.copy(lib.open(file_path).name, dest_dir)
                print(f"Moved {file_path} to {dest_dir}")
            else:
                # Recursively process subdirectories
                move_poetry_texts(file_path)
    except FileNotFoundError:
        print(f"Directory not found: {directory}")

# Move poetry texts from individual files
poetry_files = [
    "catullus.txt",
    "persius.txt",
    "prop2.txt", "prop3.txt", "prop4.txt", "propertius1.txt",
    "tibullus1.txt", "tibullus2.txt", "tibullus3.txt",
    "appvergculex.txt",
    "aus.mos.txt",
    "carm.bur.txt",
    "maximianus.txt",
    "priapea.txt",
    "valeriusflaccus1.txt", "valeriusflaccus2.txt", "valeriusflaccus3.txt",
    "valeriusflaccus4.txt", "valeriusflaccus5.txt", "valeriusflaccus6.txt",
    "valeriusflaccus7.txt", "valeriusflaccus8.txt"
]

for file_path in poetry_files:
    try:
        # Move the text file to the destination directory
        shutil.copy(lib.open(file_path).name, dest_dir)
        print(f"Moved {file_path} to {dest_dir}")
    except FileNotFoundError:
        print(f"File not found: {file_path}")

# Move poetry texts from directories
for directory in poetry_dirs:
    move_poetry_texts(directory)
```
